```python
import sys
import os

# Esto le dice a Python que también busque módulos en la carpeta src
sys.path.append(os.path.abspath("../src"))

# Ahora ya podés importar la función como si fuera un paquete
from procesar_zscore import procesar_archivo
```

```python
import pandas as pd

# Definimos los años que queremos procesar
anios = [2019, 2020, 2021, 2022, 2023]

# Procesamos cada archivo y los almacenamos en una lista
dfs = []
for anio in anios:
    file_path = f"../data/zscore_sample{anio}.csv"
    df_anio = procesar_archivo(file_path, anio)
    dfs.append(df_anio)

# Concatenamos todos los años en un solo DataFrame panel
df_panel = pd.concat(dfs, axis=0).reset_index(drop=True)

# Mostramos las primeras filas del panel
df_panel.head()
```

```python
df_panel.to_csv("../outputs/panel_zscore_limpio.csv", index=False)
```
